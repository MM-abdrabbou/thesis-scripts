```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load data from a CSV file
data = pd.read_csv('/media/rna/INIA/benchmark/Scripts/Final_Analysis/computational_time/comp_All_Aligners.csv')
```

```python
# Create a scatterplot of time vs. number of sequences
plt.scatter(data['raw_total_sequences'], data['Comp_time(s)_BWA'])
m, b = np.polyfit(data['raw_total_sequences'],data['Comp_time(s)_BWA'], 1)
plt.plot(data['raw_total_sequences'], m*data['raw_total_sequences']+b)

plt.scatter(data['raw_total_sequences'], data['Comp_time(s)_Bowtie2'])
m1, b1 = np.polyfit(data['raw_total_sequences'],data['Comp_time(s)_Bowtie2'], 1)
plt.plot(data['raw_total_sequences'], m1*data['raw_total_sequences']+b1)

# Add axis labels and a title
plt.xlabel('Number of reads [million]')
plt.ylabel('Time Taken [seconds]')
plt.title('Computation Time vs. Number of reads per cultivar')
plt.legend(['BWA-MEM', 'Bowtie2'])
#plt.savefig("Align_comp.png", format="png", dpi=1600, bbox_inches="tight")

# Display the plot
plt.show()
```
