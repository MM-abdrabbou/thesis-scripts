```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

total_variants = [2721645, 1054676, 1843068, 897776, 1279525, 831784, 1914271]
snps = [2630951, 999714, 1640828, 801910, 1152557, 746051, 1631915]
qual = [1323284, 366859, 1640802, 801896, 696846, 412960, 813315]
depth = [360813, 194455, 698552, 409252, 276320, 187156, 371356]
bi_allelic = [340490, 188837, 651714, 389454, 263437, 179833, 332731]


bcftools_bwa = [2721645, 2630951, 1323284, 360813, 340490]
bcftools_bowtie2 = [1054676, 999714, 366859, 194455, 188837]
gatk4_bwa = [1843068, 1640828, 160802, 698552, 651714]
gatk4_bowtie2 = [897776, 801910, 801896, 409252, 389454] 
octopus_bwa = [1279525, 1152557, 696846, 276320, 263437]
octopus_bowtie2 = [831784, 746051, 412960, 187156, 179833]
freebayes_bwa = [1914271, 1631915, 813315, 371356, 332731]
```

```python
# Define the labels and x axis
labels = ['BCFtools.BWA-mem', 'BCFtools.Bowtie2', 'GATK4.BWA-mem', 'GATK4.Bowtie2', 'Octopus.BWA-mem','Octopus.Bowtie2', 'FreeBayes.BWA-mem'] 
x = np.arange(len(labels))

# Define the width of each bar
width = 0.1

# Create the figure and axis objects
fig, ax = plt.subplots()

# Plot each bar with a different color
ax.bar(x - 3*width/2, total_variants, width, label='Total Variants')
ax.bar(x - width/2, snps, width, label='SNPs')
ax.bar(x + width/2, qual, width, label='Quality > 30')
ax.bar(x + 3*width/2, depth , width, label='Depth > 60')
ax.bar(x + 5*width/2, bi_allelic, width, label='Bi_allelic SNPs')


# Set the x axis ticks and labels
ax.set_xticks(x)
ax.set_xticklabels(labels)

# Set the y axis label
ax.set_ylabel('Number of variants [million]')

# Add a title and legend
ax.set_title('Variant Filtration')
#ax.legend(loc='upper center')
plt.xticks(rotation = 45)
plt.savefig("Variant_Filtration_4.png", format="png", dpi=1600, bbox_inches="tight")
# Show the plot
plt.show()
```
