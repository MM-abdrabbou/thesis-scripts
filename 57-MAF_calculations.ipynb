```python
import vcf
import matplotlib.pyplot as plt
```

```python
# Load VCF file
bcftools_bwa = vcf.Reader(open('/media/rna/INIA/benchmark/Concordance/MAF/1_biallelic/bcftools_bwa.vcf', 'r'))
bcftools_bowtie2 = vcf.Reader(open('/media/rna/INIA/benchmark/Concordance/MAF/1_biallelic/bcftools_bowtie2.vcf', 'r'))
gatk4_bwa = vcf.Reader(open('/media/rna/INIA/benchmark/Concordance/MAF/1_biallelic/gatk4_bwa.vcf', 'r'))
gatk4_bowtie2 = vcf.Reader(open('/media/rna/INIA/benchmark/Concordance/MAF/1_biallelic/gatk4_bowtie2.vcf', 'r'))
octopus_bwa = vcf.Reader(open('/media/rna/INIA/benchmark/Concordance/MAF/1_biallelic/octopus_bwa.vcf', 'r'))
octopus_bowtie2 = vcf.Reader(open('/media/rna/INIA/benchmark/Concordance/MAF/1_biallelic/octopus_bowtie2.vcf', 'r'))
freebayes_bwa = vcf.Reader(open('/media/rna/INIA/benchmark/Concordance/MAF/1_biallelic/freebayes_bwa.vcf', 'r'))
```

```python
#bcftools
bcftools_bwa = vcf.Reader(open('/media/rna/INIA/benchmark/Concordance/MAF/1_biallelic/bcftools_bwa.vcf', 'r'))
bcftools_bowtie2 = vcf.Reader(open('/media/rna/INIA/benchmark/Concordance/MAF/1_biallelic/bcftools_bowtie2.vcf'))


bcftools_bwa_maf_freq = []

for record in bcftools_bwa:
    alt_alleles = record.ALT
    
    for alt in alt_alleles:
        allele_freq = record.INFO['AF'][0] 
        
        maf = min(allele_freq, 1-allele_freq)
        bcftools_bwa_maf_freq.append(maf)

#print(bcftools_bwa_maf_freq)

bcftools_bowtie2_maf_freq = []
for record in bcftools_bowtie2:
    alt_alleles = record.ALT
    
    for alt in alt_alleles:
        allele_freq = record.INFO['AF'][0] 
        maf = min(allele_freq, 1-allele_freq)
        bcftools_bowtie2_maf_freq.append(maf)
#print(bcftools_bowtie2_maf_freq)

```

```python
fig, ax = plt.subplots()

# Plot the histograms with different colors
ax.hist(bcftools_bwa_maf_freq, bins=30, alpha=1, label='BCFtools.BWA-mem')
ax.hist(bcftools_bowtie2_maf_freq, bins=30, alpha=1, label='BCFtools.Bowtie2')

# Add a legend to the plot
ax.legend()

# Add axis labels and a title to the plot
ax.set_xlabel('Minor Allele Frequency')
ax.set_ylabel('Count')
ax.set_title('MAF distribution of BCFtools')
#plt.savefig("BCFtools_MAF_nonzero.png", format="png", dpi=1600, bbox_inches="tight")

# Show the plot
plt.show()
```
